```python
#Outlier Treatment
```

```python
def outlier_treatment(dataframe):
    Q1=dataframe.quantile(0.25)
    Q3=dataframe.quantile(0.75)
    IQR=Q3-Q1
    df_rem_outliers=[~((dataframe<(Q1-1.5*IQR))|(Q3+1.5*IQR)).any(axis=1)]
    return df_rem_outliers
```

```python
outlier_treatment(iris)
```

```python
x=iris.iloc[:,0:4].values
```

```python
y=iris.iloc[:,-1]
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_sc=sc.fit_transform(x)
x_sc
```

```python
x=pd.DataFrame(x_sc)
```

```python
x.head()
```

```python
#Imbalance Dataset
y.value_counts()
```

```python
#Balance Dataset
```

```python
#Splitting data into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.75,random_state=101)
```

```python
#Bulding Random Forest Classifier

from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
```

```python
#Tunning with Randomized Seqarch CV
```

```python
from sklearn.model_selection import RandomizedSearchCV
#Number of trees(n_estimators),max_features,max_depth,min_samples_split,min_samples_leaf in RF#parameters in RF
n_estimators=[int(x) for x in np.linspace(start=200,stop=2000,num=10)]
max_features=["auto","sqrt","log2"]
max_depth=[int(x) for x in np.linspace(10,1000,10)]   
min_samples_split=[2,5,10,14]
min_samples_leaf=[1,2,4,6,8]
```

```python
#randomized Grid Search
random_grid={'n_estimators':n_estimators,'max_features':max_features,
             'max_depth':max_depth,  'min_samples_split':min_samples_split,
             'min_samples_leaf':min_samples_leaf,
             'criterion':['entropy','gini']
              }
print(random_grid)
```

```python
rf_randomized=RandomizedSearchCV(estimator=rf,param_distributions=random_grid,n_iter=100,cv=3,verbose=2,random_state=100,n_jobs=-1)
rf_randomized.fit(x_train,y_train)
```

```python
#Evaluating
from sklearn.metrics import accuracy_score
```

```python
randomize_predictions=rf_randomized.predict(x_test)
print("Accuracy Score: ")
print(accuracy_score(y_test,randomize_predictions))

print("Tuned hyperparameters k:{}".format(rf_randomized.best_params_))
```
